<h3 align="Center"><b>Histograma</b></h3>

---

<br>

- <b>¿Como se crean?</b>

  1. <b>Determina el rango de los datos:</b><br>&emsp;<small>Identificar el valor mínimo y máximo de tus datos.</small>

  1. <b>Establece la anchura de clase:</b><br> &emsp;<small>Dividir el rango total en intervalos o clases. La anchura de clase es el ancho de cada intervalo.</small>

  1. <b>Construye los intervalos de clases:</b><br>&emsp;<small>Dividir el rango de los datos en segmentos iguales usando la anchura de clase obtenida en el paso anterior.</small>

  1. <b>Cuenta la frecuencia de datos en cada intervalo:</b><br>&emsp;<small>Para cada intervalo, contar cuántos datos caen dentro de él.</small>

  1. <b>Grafica el histograma:</b><br>&emsp;<small>Dibujar el eje horizontal (X) para representar los intervalos de clases y el eje vertical (Y) para representar las frecuencias. Cada barra del histograma tiene una anchura igual a la anchura de clase y una altura que representa la frecuencia de los datos en ese intervalo.</small>

<br>

- <b>¿Para que sirven?</b>

  - Sirven para obtener una "<b>primera vista</b>" general, o panorama, de la distribución de la población, o de la muestra, respecto a una característica, cuantitativa y continua (como la longitud o el peso).

<br>

- <b>¿En que contexto se utilizan?</b>

  - <b>Ciencias Sociales:</b><br>&emsp;<small>Permiten analizar la distribución de características cuantitativas en poblaciones, como la edad o la estatura.</small>

  - <b>Economía:</b><br>&emsp;<small>Ayudan a visualizar la distribución de variables económicas, como ingresos o salarios.</small>

  - <b>Investigación Científica:</b><br>&emsp;<small>Son útiles para representar la distribución de datos recopilados en experimentos o estudios de campo.</small>

  - <b>Control de Calidad:</b><br>&emsp;<small>En la industria, los histogramas se utilizan para analizar la variabilidad de los procesos de producción.</small>

<h3 align="Center"><b>Diagrama de tallo y hoja</b></h3>

---

<br>

- <b>¿Como se crean?</b>

  1. <b>Ordenar los datos:</b><br>&emsp;<small>Primero, se ordenan los datos de menor a mayor.</small>

  1. <b>Decidir el corte entre tallo y hoja:</b><br>&emsp;<small>Se determina cuántas cifras serán las hojas. Generalmente, la hoja es el último dígito de cada dato, y el tallo es el resto del número.</small>

  1. <b>Representar los datos:</b><br>&emsp;<small>Los tallos se colocan en la columna de la izquierda en orden ascendente, mientras que las hojas se ponen en la columna de la derecha a la altura de su tallo correspondiente y también en orden ascendente.</small>

<br>

- <b>¿Para que sirven?</b>
 
  - Permite obtener simultáneamente una distribución de frecuencias de la variable y su representación gráfica. Consiste en separar cada dato en dos partes: el “<b>tallo</b>”, que representa las cifras más significativas, y la “<b>hoja</b>”, que representa las cifras menos significativas. Este método facilita la comprensión de la forma en que los datos están distribuidos, permitiendo identificar fácilmente la asimetría y los valores atípicos, y es especialmente útil para conjuntos de datos de tamaño medio.

<br>

- <b>¿En que contexto se utilizan?</b>
  - <b>Educación:</b><br>&emsp;<small>Los diagramas de tallo y hoja son una herramienta didáctica para enseñar a los estudiantes cómo organizar y analizar datos. Por ejemplo, se pueden utilizar para representar la edad de un grupo de personas, la altura de estudiantes, o incluso los resultados de una prueba.</small>

  - <b>Investigación científica:</b><br>&emsp;<small>Se pueden usar para analizar y presentar datos recopilados durante experimentos. Por ejemplo, un científico podría usar un diagrama de tallo y hoja para evaluar la distribución del porcentaje de grasa en muestras de salsa embotellada.</small>

<h3 align="Center"><b>Diagrama de caja</b></h3>

---

<br>

- <b>¿Como se crean?</b>

  1. <b>Ordenar los datos:</b><br>&emsp;<small>Primero, se ordenan los datos de menor a mayor.</small>

  1. <b>Calcular los cuartiles:</b><br>&emsp;<small>Se determinan los tres cuartiles (Q1, Q2, Q3). Q1 es el primer cuartil o percentil 25, Q2 es la mediana o percentil 50, y Q3 es el tercer cuartil o percentil 75.</small>

  1. <b>Calcular el rango intercuartílico (IQR):</b><br>&emsp;<small>El IQR se calcula restando el primer cuartil (Q1) del tercer cuartil (Q3). Esto da una idea de la dispersión de los datos.</small>

  1. <b>Determinar los límites:</b><br>&emsp;<small>Se calculan los límites inferior y superior para identificar los valores atípicos. Los límites inferiores y superiores se calculan como Q1 - 1.5 * IQR y Q3 + 1.5 * IQR, respectivamente.</small>

  1. <b>Identificar los valores atípicos:</b><br>&emsp;<small>Los valores que caen por debajo del límite inferior o por encima del límite superior son considerados valores atípicos.</small>

  1. <b>Dibujar el diagrama de caja:</b><small>
    - &emsp;Se dibuja una caja que va desde Q1 hasta Q3.
    - &emsp;Se marca la mediana (Q2) dentro de la caja.
    - &emsp;Se dibujan las líneas (bigotes) que se extienden desde los extremos de la caja hasta los valores mínimo y máximo que no son atípicos.
    - &emsp;Si existen valores atípicos, se representan como puntos individuales fuera de los bigotes.</small>

<br>

- <b>¿Para que sirven?</b>

  - Comparar el intervalo y la distribución de conjuntos de datos numéricos.
  - Identificar rápidamente la dispersión y la simetría de los datos.
  - Detectar posibles valores atípicos en los datos.
  - Comparar dos o más conjuntos de datos de manera visual.

<br>

- <b>¿En que contexto se utilizan?</b>

  - <b>Análisis de Datos:</b><br>&emsp;<small>Ayudan a comprender la distribución de los datos, identificar valores atípicos y realizar comparaciones entre diferentes grupos de datos.</small>

  - <b>Estadística Descriptiva:</b><br>&emsp;<small>Proporcionan una representación visual concisa de la tendencia central, la dispersión y la asimetría de los datos.</small>

  - <b>Control de Calidad:</b><br>&emsp;<small>Se utilizan para monitorear procesos y detectar variaciones o valores atípicos que podrían indicar problemas de calidad.</small>

  - <b>Investigación Médica:</b><br>&emsp;<small>Se emplean para analizar datos de pacientes, identificar tendencias y comparar diferentes grupos de tratamiento.</small>

  - <b>Finanzas:</b><br>&emsp;<small>Se utilizan para analizar el rendimiento de las inversiones, identificar valores atípicos y comprender la volatilidad del mercado.</small>

<h3 align="Center"><b>Diagrama de puntos</b></h3>

---

<br>

- <b>¿Como se crean?</b>

  - <b>Reunir los datos:</b><br>&emsp;<small>Estos datos pueden ser de cualquier tipo, pero generalmente se trata de una variable cuantitativa.</small>
  
  - <b>Organizar los datos:</b><br>&emsp;<small>Organizar los datos de menor a mayor para facilitar su interpretación.</small>
  
  - <b>Crear una línea horizontal:</b><br>&emsp;<small>Dibujar una línea horizontal que represente la recta numérica de los datos. Esta línea debe comenzar con el valor mínimo y terminar con el valor máximo de los datos.</small>
  
  - <b>Etiquetar la línea:</b><br>&emsp;<small>Etiquetar la línea horizontal con los valores correspondientes a los datos</small>
  
  - <b>Representar los datos:</b><br>&emsp;<small>Para cada valor en el conjunto de datos, coloca un punto sobre la línea horizontal en la posición correspondiente al valor. Si un valor se repite, coloca tantos puntos como veces se repite ese valor.</small>

<br>

- <b>¿Para que sirven?</b>

  - Sirve para representar gráficamente la relación entre dos variables. Este tipo de diagrama es útil para identificar patrones, correlaciones y tendencias entre los datos. Algunas de las principales funciones del diagrama de puntos son:
    - <small><b>Identificar correlaciones.</b></small>
    - <small><b>Detectar valores atípicos.</b></small>
    - <small><b>Visualizar distribuciones.</b></small>
    - <small><b>Hacer predicciones.</b></small>
    - <small><b>Comparar conjuntos de datos.</b></small>

<br>

- <b>¿En que contexto se utilizan?</b>

  - <b>Análisis de datos:</b><br>&emsp;<small>Ayudan a los analistas a identificar clustering, distribuciones superpuestas y dispersión en los datos, proporcionando información sobre los datos y permitiendo realizar cambios en las variables explicativas para crear un modelo más eficaz.</small>

  - <b>Proyecciones económicas:</b><br>&emsp;<small>El gráfico de puntos más famoso es el de las previsiones de tipos de interés de la Reserva Federal que se publican cada trimestre, donde se muestran las proyecciones de los miembros del FOMC sobre los tipos de interés futuros.</small>

  - <b>Educación:</b><br>&emsp;<small>Se utilizan en la enseñanza para organizar datos de una manera fácil y rápida, como en el ejemplo de la cantidad de libros leídos por estudiantes durante el verano.</small>

  - <b>Investigación:</b><br>&emsp;<small>Ayudan a descubrir patrones en los datos, como en el caso de las emisiones de gases de efecto invernadero a lo largo del tiempo, donde los puntos pueden indicar variaciones anuales en diferentes países.</small>
